# Image Input

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from ipywidgets import FileUpload
from IPython.display import display

uploader=FileUpload(accept='.png',multiple=False)
display(uploader)
```

```python
import base64

uploaded_file=uploader.value[0]
content_mv=uploaded_file["content"]
img_bytes=bytes(content_mv)
img_b64=base64.b64encode(img_bytes).decode("utf-8")
```

```python
agent=create_agent(model='gpt-5-nano')

multimodal_question=HumanMessage(content=[
    {"type":"text","text":"tell me more about the image"},
    {"type":"image","base64":img_b64,"mime_type":"image/png"}
    ])

response=agent.invoke({'messages':[multimodal_question]})
print(response['messages'][-1].content)
```

# Audio input

```python
import sounddevice as sd
from scipy.io.wavfile import write
import base64
import io
import time
from tqdm import tqdm

#Recording settings
duration=5 #seconds
sample_rate=44100
print("Recording...")
audio=sd.rec(int(duration*sample_rate),samplerate=sample_rate,channels=1)
#Progress bar
for _ in tqdm(range(duration*10)):
    time.sleep(0.1)

sd.wait()
print("Done.")

#Write Wav to add in-memory buffer
buf=io.BytesIO()
write(buf,sample_rate,audio)
wav_bytes=buf.getvalue()
aud_b4=base64.b64encode(wav_bytes).decode("utf-8")
```

```python
agent=create_agent(model='gpt-4o-audio-preview',)

multimodal_question=HumanMessage(content=[
    {"type":"text","text":"do what the audio says"},
    {"type":"audio","base64":aud_b4,"mime_type":"audio/wav"}
])
response=agent.invoke({"messages":[multimodal_question]})

print(response['messages'][-1].content)
```

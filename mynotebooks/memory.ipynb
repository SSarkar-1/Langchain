```python
from dotenv import load_dotenv
load_dotenv()
```

# No memory

```python
from langchain.agents import create_agent
agent=create_agent(model="gpt-5-nano")
```

```python
from langchain.messages import HumanMessage
question=HumanMessage(content="I am Shubham Sarkar.My good wife is the beautiful Trisha Debnath Sarkar.I am a Multibillonaire and own more then 50 properties and farmhouses.I have a lambo")
response=agent.invoke({"messages":[question]})
print(response['messages'][-1].content)
```

```python
question=HumanMessage(content="Who is my wife?")
response=agent.invoke({'messages':[question]})
```

```python
print(response['messages'][-1].content)
```

# Memory

```python
from langgraph.checkpoint.memory import InMemorySaver

agent=create_agent(model='gpt-5-nano',checkpointer=InMemorySaver(),)
```

```python
question=HumanMessage(content="I am Shubham Sarkar.My good wife is the beautiful Trisha Debnath Sarkar.I am a Multibillonaire and own more then 50 properties and farmhouses.I have a lambo")
config={"configurable":{"thread_id":"1"}}
response=agent.invoke({"messages":[question]},config)
```

```python
print(response['messages'][-1].content)
```

```python
question=HumanMessage(content="Who is my wife?")
response=agent.invoke({"messages":[question]},config,)
print(response['messages'][-1].content)
```

```python
question=HumanMessage(content="how does she look?")
response=agent.invoke({"messages":[question]},config,)
print(response['messages'][-1].content)
```

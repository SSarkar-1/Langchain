```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.agents import AgentState

class CustomState(AgentState):
    favourite_food: str
```

# Write to state

```python
from langchain.tools import tool,ToolRuntime
from langgraph.types import Command
from langchain.messages import HumanMessage,ToolMessage

@tool
def update_favourite_food(favourite_food:str,runtime:ToolRuntime):
    """Update the favourite food of the user in the state once they have revealed it"""
    return Command(update={
        "favourite_food":favourite_food,
        "messages":[ToolMessage("Successfully Updated favourite Food",tool_call_id=runtime.tool_call_id)]
    })
```

```python
from langchain.agents import create_agent
from langgraph.checkpoint.memory import InMemorySaver

agent=create_agent(model="gpt-5-nano",tools=[update_favourite_food],
                   checkpointer=InMemorySaver(),state_schema=CustomState)
```

```python
response=agent.invoke({"messages":[HumanMessage(content="My favourite food is biryani")]},
                      {"configurable":{"thread_id":"1"}})
```

```python
from pprint import pprint
pprint(response)
```

```python
response=agent.invoke({"messages":[HumanMessage(content="Hi how are you")], "favourite_food":"biryani"},
                     { "configurable":{"thread_id":"10"}})

pprint(response)
```

```python
response = agent.invoke(
    { "messages": [HumanMessage(content="What's my favourite food?")]},
    {"configurable": {"thread_id": "1"}}
)

pprint(response)
```

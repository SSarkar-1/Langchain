```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain.tools import tool

@tool
def calculate_square_root(x):
    """Calculate the square root of a number"""
    return x**0.5
```

```python
calculate_square_root.invoke({"x":225})
```

# Adding tool to agents

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage

agent=create_agent(model="gpt-5-nano",tools=[calculate_square_root],system_prompt="You are a maths wizard> use tools to calculate square and square root of any number")
```

```python
question=HumanMessage(content="What is the square root of 625?")
response=agent.invoke({"messages":[question]})
print(response['messages'][-1].content)
```

```python
from pprint import pprint
pprint(response['messages'])
```

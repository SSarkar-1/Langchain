```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from dataclasses import dataclass

@dataclass
class FoodContext:
    favourite_food: str="biryani"
    least_favourite_food: str="macher jhol"
```

```python
from langchain.agents import create_agent

agent=create_agent(model="gpt-5-nano",
                   context_schema=FoodContext)
```

```python
from langchain.messages import HumanMessage

response=agent.invoke({"messages":[HumanMessage(content="What is my favourite food")]})
print(response)
```

```python
from pprint import pprint

pprint(response)
```

# Accessing Context

```python
from langchain.tools import tool, ToolRuntime

@tool
def get_favourite_food(runtime):
    """Get favourite food of the user"""
    return runtime.context.favourite_food

@tool
def get_least_favourite_food(runtime):
    """Get least favourite food of the user"""
    return runtime.context.least_favourite_food
```

```python
agent=create_agent(model="gpt-5-nano",
                   tools=[get_least_favourite_food,get_favourite_food],
                   context_schema=FoodContext)
```

```python
response=agent.invoke({"messages":[HumanMessage(content="What is my favourite food")]},
                      context=FoodContext())
```

```python
pprint(response)
```

```python
response=agent.invoke(
    {"messages":[HumanMessage(content="What is my least favourite food?")]},
    context=FoodContext(least_favourite_food='macher jhol')
)
pprint(response)
```
